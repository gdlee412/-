```python
sales = [25, 35, 15]
menu = ['아메리카노', '카페라테', '마키아또']
plt.title("Bar Chart")
plt.xlabel("판매수")
plt.ylabel("메뉴")
plt.barh(menu,sales)
plt.show()
```

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels)
plt.show()
```

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0.05, 0.05, 0.05, 0.05]
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
plt.pie(ratio, labels=labels, autopct='%.1f%%', explode=explode, shadow=True, colors=colors)
plt.show()
```

```python
plt.subplot(2,2,1)
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0.05, 0.05, 0.05, 0.05]
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
plt.pie(ratio, labels=labels, explode=explode, shadow=True, colors=colors)
plt.title("data 1")

plt.subplot(2,2,2)
sales = [25, 35, 15]
menu = ['아메리카노', '카페라테', '마키아또']
plt.barh(menu,sales)
plt.title("data 2")

plt.subplot(2,2,3)
x = np.random.randn(1000)
plt.hist(x, bins=100)
plt.title("data 3")

plt.subplot(2,2,4)
data = np.random.normal(0, 2.0, 1000)
data_b = np.random.normal(-3.0, 1.5, 1000)
data_c = np.random.normal(1.2, 1.5, 1000)
plt.boxplot([data, data_b, data_c])
plt.title("data 4")
plt.tight_layout()
plt.show()
```
